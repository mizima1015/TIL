# 문제: 두 수의 합

주어진 두 정수 배열 nums1과 nums2가 있습니다. 각 배열에서 하나씩 숫자를 선택하여 두 숫자의 합이 주어진 타겟 숫자 target이 되도록 하세요. 모든 가능한 조합을 찾아 반환하는 함수를 작성하세요.

입력:

nums1: 정수 배열 (예: [1, 2, 3])

nums2: 정수 배열 (예: [3, 4, 5])

target: 정수 (예: 5)

출력:

가능한 모든 조합을 담은 리스트. 각 조합은 두 요소를 가진 튜플로 구성됩니다.

(예: [(1, 4), (2, 3)])

## 조건

각 배열의 길이는 1 이상, 1000 이하입니다.

각 배열의 원소는 -1000 이상, 1000 이하의 정수입니다.

출력 조합은 순서에 상관없이 반환하시면 됩니다.

```python
# 예시

# 입력
nums1 = [1, 2, 3]
nums2 = [3, 4, 5]
target = 5

# 출력
[(1, 4), (2, 3)]

# 입력
nums1 = [0, -1, 2]
nums2 = [1, 2, 3]
target = 3

# 출력
[(0, 3), (2, 1)]
```

```python
# 풀이
nums1 = [1, 2, 3]
nums2 = [3, 4, 5]
target = 5

tlist = []

for i in range(len(nums1)):
  for j in range(len(nums2)):
    t = nums1[i] + nums2[j]
    if target == t:
      tlist.append((nums1[i],nums2[j]))


print(tlist)
```

* 이중 for 루프를 사용하는 대신 더 효율적인 방법으로 구현하는법
- set

```python
def find_pairs(nums1: list[int], nums2: list[int], target: int) -> list[tuple[int, int]]:
    result = []
    nums2_set = set(nums2)  # nums2를 집합으로 변환하여 검색 속도 향상

    for num1 in nums1:
        complement = target - num1  # 타겟을 만들기 위한 보수 값
        if complement in nums2_set:
            result.append((num1, complement))

    return result

# 테스트
nums1 = [1, 2, 3]
nums2 = [3, 4, 5]
target = 5
print(find_pairs(nums1, nums2, target))
```
